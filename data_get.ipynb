```python
#in clubname, replace ' FC' with nothing
prem_champ_df['club_name'] = prem_champ_df['club_name'].str.replace(' FC', '')
```

```python
#group by club_name, year and sum fee_cleaned
prem_champ_df = prem_champ_df.groupby(['club_name', 'year']).sum('fee_cleaned').reset_index()
prem_champ_df
```

```python
#give me unique club names in prem_champ_df
prem_champ_df['club_name'].unique()
```

```python
#remove the age column
prem_champ_df = prem_champ_df.drop(columns=['age'])
```

```python
#unmelt this dataframe
prem_champ_df = prem_champ_df.pivot(index='club_name', columns='year', values='fee_cleaned').reset_index()
prem_champ_df
```
